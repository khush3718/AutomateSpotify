```python
#Get the user's playlists

import pandas as pd

source_playlistID = '37i9dQZF1E36aG4ozWX8uv'
source_playlist = sp.user_playlist(username, source_playlistID)
source_tracks = source_playlist["tracks"]
source_songs = source_tracks["items"]

tracks_ids = [] # To store the ids of the tracks
tracks_names = [] # To store the names of the tracks

for i in range(0, len(source_songs)):
    if source_songs[i]['track']['id'] != None:                    # Removes the local tracks in your playlist if there is any
        tracks_ids.append(source_songs[i]['track']['id'])
        tracks_names.append(source_songs[i]['track']['name'])

features = [] # To store the features of the tracks
for i in range(0,len(tracks_ids)):
    audio_features = sp.audio_features(tracks_ids[i])
    for track in audio_features:
        features.append(track)

playlist_df = pd.DataFrame(features, index = tracks_names) # Create dataframe from features
```

```python
playlist_df=playlist_df[["id", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness",  "key", "liveness",
                         "loudness", "mode", "speechiness", "tempo", "valence"]]
playlist_df.head()
```

```python
# Generate a new column for recommended tracks
# Set recommendation limit to the Playlist Length per track

playlist_df['recommended_tracks'] =  playlist_df['id'].apply(lambda x: [(track['id']) for track in sp.recommendations(seed_tracks = [x], limit = len(playlist_df))['tracks']])
```

```python
# take a look at the recommended tracks for the first song in the playlist

for track in sp.tracks(playlist_df['recommended_tracks'][0])['tracks']:
    print(track['name'] + ' - ' + track['artists'][0]['name'] + ' - ' + track['external_urls']['spotify'])
```
